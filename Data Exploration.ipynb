## Simple visualization
### Burglary distribution per time period

```python
#Burglary distribution per month and year
sns.set_style("whitegrid")
df_grouped = df.groupby(['Year', 'Month_Num'], as_index=False)['Burglary_Count'].sum()

df_grouped['month_name'] = df_grouped['Month_Num'].apply(lambda x: calendar.month_abbr[x])

pivot_df = df_grouped.pivot(index='month_name', columns='Year', values='Burglary_Count')

month_order = list(calendar.month_abbr)[1:]  # ['Jan', 'Feb', ..., 'Dec']
pivot_df = pivot_df.reindex(month_order).dropna(how='all')

fig, ax = plt.subplots(figsize=(16, 8))
pivot_df.plot(kind='bar', ax=ax, width=0.8, colormap='tab20')

ax.set_title('Monthly Burglary Counts per Month by Year', fontsize=18)
ax.set_xlabel('Month', fontsize=14)
ax.set_ylabel('Burglary Count', fontsize=14)

ax.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

plt.tight_layout()
plt.show()
```

```python
#Average number of burglary per month
monthly_avg = df.groupby('Month_Num')['Burglary_Count'].mean().reset_index()
monthly_avg
```

```python
#Bar chart of average number of burglary per month
plt.figure(figsize=(8, 6))
plt.bar(monthly_avg['Month_Num'], monthly_avg['Burglary_Count'], color='skyblue')
plt.xlabel('Month Number')
plt.ylabel('Average Burglary Count')
plt.title('Average Number of Burglaries Per Month')
plt.xticks(monthly_avg['Month_Num'])
plt.show()
```

```python
#Scatter plot between Burglary count and daylight hours of month
plt.figure(figsize=(8, 6))
plt.scatter(df['Precise_Daylight_Hours'], df['Burglary_Count'], color='red', alpha=0.7)
plt.xlabel('Daylight Hours')
plt.ylabel('Burglary Count')
plt.title('Burglary Count vs. Daylight Hours')
plt.grid(True)
plt.show()
```

### Socio-economical distribution per LSOA

```python
#Stacked bar chart of industry distribution in each LSOA
industry_columns = [
    'Ind_Agriculture',
    'Ind_Manufacturing',
    'Does not apply', 
    'Ind_Construction',
    'Ind_Distribution',
    'Ind_Transport'
]

df_industry = df[['LSOA code'] + industry_columns]
df_grouped = df_industry.groupby('LSOA code').sum()

# Plot the stacked bar chart
df_grouped.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.ylabel('Number of People')
plt.xlabel('LSOA Code')
plt.title('People Working in Different Industries by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Stacked bar chart of industry distribution percentage in each LSOA
df_percent = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

df_percent.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.ylabel('Percentage of Workforce')
plt.xlabel('LSOA Code')
plt.title('Percentage of People Working in Different Industries by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Stacked bar chart of house type percentage in each LSOA
house_type = [
    'Flat, maisonette or apartment',
    'A caravan or other mobile or temporary structure',
    'Whole house or bungalow: Detached', 
    'Whole house or bungalow: Semi-detached',
    'Whole house or bungalow: Terraced',
]

df_house_type= df[['LSOA code'] + house_type]
df_grouped = df_house_type.groupby('LSOA code').sum()

df_percent = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

df_percent.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.ylabel('Percentage of House Type')
plt.xlabel('LSOA Code')
plt.title('Percentage of House Type Distribution by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='House Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Stacked bar chart of room density percentage in each LSOA
room_density= [
    'Over 1.0 and up to 1.5 people per room',
    'Over 1.0 and up to 1.5 people per room',
    'Over 1.5 people per room', 
    'Up to and including 0.5 people per room',
]

df_room_density = df[['LSOA code'] + room_density]
df_grouped = df_room_density.groupby('LSOA code').sum()

df_percent = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

df_percent.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.ylabel('Percentage of Room Density')
plt.xlabel('LSOA Code')
plt.title('Percentage of Room Density Distribution by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Room Density', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Stacked bar chart of ethnicity percentage in each LSOA
ethnicity= [
    'Asian, Asian British or Asian Welsh: Bangladeshi',
    'Asian, Asian British or Asian Welsh: Chinese',
    'Asian, Asian British or Asian Welsh: Indian', 
    'Asian, Asian British or Asian Welsh: Other Asian',
    'Asian, Asian British or Asian Welsh: Pakistani',
    'Black, Black British, Black Welsh, Caribbean or African: African',
    'Black, Black British, Black Welsh, Caribbean or African: Caribbean',
    'Black, Black British, Black Welsh, Caribbean or African: Other Black',
    'Mixed or Multiple ethnic groups: Other Mixed or Multiple ethnic groups',
    'Mixed or Multiple ethnic groups: White and Asian',
    'Mixed or Multiple ethnic groups: White and Black African',
    'Mixed or Multiple ethnic groups: White and Black Caribbean',
    'Other ethnic group: Any other ethnic group',
    'Other ethnic group: Arab',
    'White: English, Welsh, Scottish, Northern Irish or British',
    'White: Gypsy or Irish Traveller',
    'White: Irish',
    'White: Other White',
    'White: Roma'
]

df_ethnicity = df[['LSOA code'] + ethnicity]
df_grouped = df_ethnicity.groupby('LSOA code').sum()

df_percent = df_grouped.div(df_grouped.sum(axis=1), axis=0) * 100

df_percent.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')
plt.ylabel('Percentage of Ethnicity')
plt.xlabel('LSOA Code')
plt.title('Percentage of Ethnicity Distribution by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Ethnicity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
broad_ethnicity_map = {eth: eth.split(":")[0] for eth in ethnicity}

df_ethnicity = df[['LSOA code'] + ethnicity].copy()

df_ethnicity_renamed = df_ethnicity.rename(columns=broad_ethnicity_map)

df_broad = df_ethnicity_renamed.groupby('LSOA code', axis=0).sum()
df_broad_grouped = df_broad.groupby(df_broad.columns, axis=1).sum()

df_percent = df_broad_grouped.div(df_broad_grouped.sum(axis=1), axis=0) * 100

df_percent.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab10')
plt.ylabel('Percentage of grouped ethnicity')
plt.xlabel('LSOA Code')
plt.title('Percentage of Ethnicity Group Distribution by LSOA')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Ethnicity Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Change in Socio-economical data in London

```python
#Industry spead change over time
valid_cols = ['Year'] + [col for col in industry_columns if col in df.columns]
df_subset = df[valid_cols]

df_grouped_year = df_subset.groupby('Year').sum()

df_percent_year = df_grouped_year.div(df_grouped_year.sum(axis=1), axis=0) * 100

df_percent_year.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.ylabel('Percentage of Workforce')
plt.xlabel('Year')
plt.title('Industry Spread Over Time in Percentage')
plt.xticks(rotation=0)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#House type change over time
valid_cols = ['Year'] + [col for col in house_type if col in df.columns]
df_subset = df[valid_cols]

df_grouped_year = df_subset.groupby('Year').sum()

df_percent_year = df_grouped_year.div(df_grouped_year.sum(axis=1), axis=0) * 100

df_percent_year.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.ylabel('Percentage of House type')
plt.xlabel('Year')
plt.title('Change in Housing type Over Time in Percentage')
plt.xticks(rotation=0)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
