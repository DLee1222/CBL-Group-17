```python
import pandas as pd
```

```python
hotspots = pd.read_csv("hot-spot.csv")
burglary = pd.read_csv("merged_burglary_scores_2010_2025.csv")
LSOA = pd.read_csv("UK_LSOA_data.csv")
hotspot_50 = pd.read_csv("hot_spots_top50%.csv")
```

```python
filtered_burglary = burglary[burglary['LSOA code'].isin(hotspots['LSOA code'])]
filtered_burglary
```

```python
burglary_not_in_hotspots = burglary[~burglary['LSOA code'].isin(hotspots['LSOA code'])]
unique_lsoas = burglary_not_in_hotspots['LSOA code'].unique()
unique_lsoas, len(unique_lsoas)
```

```python
active_lsoas = LSOA.loc[LSOA['active'] == True, 'code']

activate_lsoa = filtered_burglary[filtered_burglary['LSOA code'].isin(active_lsoas)]
activate_lsoa
```

```python
dataset_ML = pd.merge(
    activate_lsoa,
    hotspots[['LSOA code', 'longitude', 'latitude']],
    on='LSOA code',
    how='left'
)
dataset_ML
```

```python
missing_lsoa_codes = dataset_ML[dataset_ML['IMD SCORE'].isna()]['LSOA code'].unique()
missing_lsoa_codes
```

```python
missing_years = dataset_ML[dataset_ML['IMD SCORE'].isna()]['year'].unique()
missing_years
```

```python
data = {
    "IMD SCORE": [15.903667, 15.945980, 22.513889],
    "EMPLOYMENT SCORE": [0.040835, 0.053574, 0.089941],
    "EDUCATION SKILLS AND TRAINING SCORE": [6.857291, 4.773817, 13.001624],
    "INCOME SCORE": [0.072916, 0.073980, 0.140992],
    "BARRIERS TO HOUSING AND SERVICES SCORE": [33.790574, 32.065924, 32.181404],
    "LIVING ENVIRONMENT SCORE": [52.884274, 55.008751, 30.846190]
}
index = ["2010-2014", "2015-2018", "2019-2025"]
average_score = pd.DataFrame(data, index=index)

average_score.index.name = "Year_Group"
average_score
```

```python
fill_values = average_score.loc["2019-2025"]

for column in fill_values.index:
    if column in dataset_ML.columns:
        dataset_ML[column] =dataset_ML[column].fillna(fill_values[column])
```

```python
missing_values = dataset_ML.isnull().sum()
missing_values
```

```python
hotspot_ML = dataset_ML[dataset_ML['LSOA code'].isin(hotspot_50['LSOA code'])]
hotspot_ML
```

```python
missing_values_hotspot = hotspot_ML.isnull().sum()
missing_values_hotspot
```

```python
dataset_ML.to_csv('dataset_ML.csv', index=False)
hotspot_ML.to_csv('hotspot_ML.csv', index=False)
```
